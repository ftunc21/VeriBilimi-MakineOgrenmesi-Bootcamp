```python
#ordinal tanımlama 
# nominal , bir kişinin cinsiyeti, etnik kökeni, saç rengi vb. nominal bir ölçek için veri olarak kabul edilir.
#ordinal , bilgilerin belirli bir sıraya göre, yani birbirleriyle karşılaştırmalı olarak düzenlenmesini ve her bir parametrenin (değişkenin) “sıralanmasını” içerir. 
from pandas.api.types import CategoricalDtype
```

```python
#company_size sütununu kategorik bir veri türüne dönüştürür.
#CategoricalDtype(ordered=True) ifadesi, bu kategorik veri türünün sıralanabilir (ordered) olduğunu belirtir. Yani, bu sütunun değerleri arasında bir sıralama yapılabilir.
df.company_size = df.company_size.astype(CategoricalDtype(ordered = True))
```

```python
#sıralama yanlış yapıldığı için biz manuel olarak sıralama bilgisini veriyoruz
categories = ["S","M","L"]
```

```python
df.company_size = df.company_size.astype(CategoricalDtype(categories = categories, ordered = True))
```

# #Veri Görselleştirme

```python
#Bütün meslekleri bir arada bir grafikte gözlemleyelim
plt.figure(figsize= (25,10))
sns.countplot(x="job_title", data=df,palette = "hls")
plt.ylabel("Toplam", size=20)
plt.xlabel('Meslek', size=20)
plt.xticks(rotation=90)
plt.title("Mesleklerin Toplam Sayıları", size=20);

```

```python
#Meslek — ortalama maaş grafiği
job_salary_usd = df[['job_title', 'salary_in_usd']]
job_salary_usd  = df.groupby('job_title').agg({'salary_in_usd': 'mean'}).sort_values('salary_in_usd', ascending=False).reset_index().head(50)
fig = plt.figure(figsize=(14,12))
sns.barplot(x='job_title',
            y='salary_in_usd',
            data=job_salary_usd
            )
plt.ylabel("Ortalama Maaş", size=20)
plt.xlabel('Meslek', size=20)
plt.xticks(rotation=90)
plt.title("Mesleklerin Ortalama Maaşları (USD)", size=25);
```
